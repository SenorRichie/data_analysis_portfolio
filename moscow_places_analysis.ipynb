**hours - вероятно к пропускам привело то, что владельцы заведений в сервисе карт не указали время работы. Заменим пропуски на unknown:**

**price - мы не можем дать категорию оценки, так-как нет данных, за счет которых мы могли бы её категоризовать. Пропуски возникли вероятно из-за того, что система сервиса карт не смогла дать категорию. Заменяем также на unknown:**

**avg_bill - тут похожая ситуация, что система сервиса не смогла сделать записи. Сделать мы ничего не сможем, кроме того как поставить значение unknown:**

```python
places['avg_bill'] = places['avg_bill'].fillna('unknown')
```

**middle_avg_bill - по описанию столбца, если в avg_bill есть подстрока "Средний счёт", то данные должны быть заполнены. Проверим это:**

```python
(places.loc[places['avg_bill'].str.contains('Средний счёт')])[['avg_bill', 'middle_avg_bill']].isna().sum()
```

**Всё верно, значит пропуски в middle_avg_bill означают, что в avg_bill записан не средний счёт, а другое значение. Поставим на пропуски заглушку -1:**

```python
places['middle_avg_bill'] = places['middle_avg_bill'].fillna(-1)
```

**Проверим, нет ли аномальных значений в среднем чеке:**

```python
def anomalies_test(column, title, y_label):
    '''
    Функция принимает название колонки, заголовок
    графика и подпись оси Y.
    Выводит график разброса, исключает в графике
    значения -1 по оси Y.
    '''
    # Фиксируем маску с заглушками для их удаления в графике
    mask = places.loc[places[column] == -1]
    
    # Строим график разброса
    plt.figure(figsize=(18, 8))
    ax = sns.stripplot(y=column, data=places.drop(index=mask.index))
    ax.set_title(title)
    ax.set_xlabel('Заведения')
    ax.set_ylabel(y_label)
    plt.show()

anomalies_test('middle_avg_bill', 'Разброс среднего чека по всем заведениям', 'Средний чек')
```

**Есть слишком выбивающий средний чек свыше 35000, убираем такое заведение:**

### Создайте столбец street с названиями улиц из столбца с адресом.

```python
def street_include(address):
    '''
    Функция возвращает вторую запись после запятой
    в ячейке. Применимо для извлечения названия улицы
    из полного адреса.
    '''
    return (address.split(', '))[1]

places['street'] = places['address'].apply(street_include)
```

### Создайте столбец is_24_7 с обозначением, что заведение работает ежедневно и круглосуточно.
- **логическое значение True — если заведение работает ежедневно и круглосуточно;**
- **логическое значение False — в противоположном случае.**

```python
def is_24_7_include(hours):
    '''
    Функция проверяет наличие подстрок и возвращает True,
    если они есть. В противном случае возвращает False.
    Подходит для категоризации заведении, работают ли они
    ежедневно и круглосуточно.
    '''
    if 'руглосут' in hours and 'жедневн' in hours:
        return True
    else:
        return False

places['is_24_7'] = places['hours'].apply(is_24_7_include)
```
